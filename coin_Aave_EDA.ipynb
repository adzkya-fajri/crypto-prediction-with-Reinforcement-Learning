```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18, 12)) 

# 1. SPX - Open
plt.subplot(2, 3, 1)
sns.boxplot(y=df['Open'], color='skyblue')
plt.title("Boxplot of SPX (Open)")

# 2. GLD - Close
plt.subplot(2, 3, 2)
sns.boxplot(y=df['Close'], color='pink')
plt.title("Boxplot of GLD (Close)")

# 3. USO - High
plt.subplot(2, 3, 3)
sns.boxplot(y=df['High'], color='#e78ac3')
plt.title("Boxplot of USO (High)")

# 4. SLV - Low
plt.subplot(2, 3, 4)
sns.boxplot(y=df['Low'], color='#ffd92f')
plt.title("Boxplot of SLV (Low)")

# 5. Volume (Fixed index to 5 and fixed spelling)
plt.subplot(2, 3, 5)
sns.boxplot(y=df['Volume'], color='#8da0cb') 
plt.title("Boxplot of Volume")

# 6. EUR/USD - Marketcap (Fixed index to 6)
plt.subplot(2, 3, 6)
sns.boxplot(y=df['Marketcap'], color='lightgreen')
plt.title("Boxplot of EUR/USD (Marketcap)")

plt.tight_layout()
plt.show()
```

```python
# 1. Menghitung matriks korelasi
# Pastikan menggunakan kolom yang sudah ditransformasi (log/winsorized)
corr_matrix = df[['Open', 'Close', 'High', 'Low', 'Volume', 'Marketcap']].corr()

# 2. Membuat Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Heatmap Korelasi Antar Variabel (Setelah Handling Outlier)', fontsize=15)
plt.show()
```
